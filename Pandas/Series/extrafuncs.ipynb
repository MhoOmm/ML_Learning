```python
import pandas as pd
import numpy as np
subs = pd.read_csv("subs.csv").squeeze("columns")

total_sub = subs.sum()
prod = subs.prod() # product of values
print(total_sub)
```

```python
# statistical methods
print(subs.mean())
print(subs.median())
print(subs.mode())
print(subs.std())
print(subs.var())
```

```python
# min max
print(subs.min())
print(subs.max())

# describe
print(subs.describe())
```
