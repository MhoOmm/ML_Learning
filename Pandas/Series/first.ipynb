```python
import numpy as np
import pandas as pd

# making a series

# string
country = ["India","USA","Japan","France"]
pd.Series(country)

# integer 
runs = [23,4,67,98,89,101]
pd.Series(runs)


# giving a custom index
marks = [67,78,99,75]
subjects = ["Hindi","Maths","CS","SCience"]

pd.Series(marks,index=subjects)

# giving a  name to the the series
marks = [67,78,99,75]
subjects = ["Hindi","Maths","CS","SCience"]

marks = pd.Series(marks,index=subjects,name = "Marks of student")
print(marks)

```

```python
# using a dictionary to make a series
dict1 = {"Science":89,"Maths":87,"CS":99,"English":81}
marks = pd.Series(dict1,name="MArks of a student")
print(marks)
```

```python
# attributes of series
dict1 = {"Science":89,"Maths":87,"CS":99,"English":81}
marks = pd.Series(dict1,name="MArks of a student")

# size

print(marks.size)

# dtype
print(marks.dtype)

# name attribute
print(marks.name)

# is unique
print(marks.is_unique)

# index
print(marks.index)

# values
print(marks.values)
```
