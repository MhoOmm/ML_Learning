```python
# agg method
import pandas as pd
import numpy as np
movies = pd.read_csv("imdb-top-1000.csv")
genres = movies.groupby('Genre')
# giving as a dictionary
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# agg method
import pandas as pd
import numpy as np
movies = pd.read_csv("imdb-top-1000.csv")
# genres = movies.groupby('Genre')
# genres.agg(['min','max','mean']) #har colum ke upar ye operation(list ke lag jayeng )
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
for group,data in genres:
    print(data)
```

```python
# highest rated movie in each genre
for group,data in genres:
    print(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
```

```python
# apply function 
# general procedure -> split(apply)combine
genres.apply(min)
```

```python
# number of movies starting with a in each genre
def finda(genre):
    return genre['Series_Title'].str.startswith('A').sum()

genres.apply(finda)
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group

genres.apply(rank_movie)
```

```python
def apply_normal(group):
    minval = group['IMDB_Rating'].min()
    maxval = group['IMDB_Rating'].max()
    val = group['IMDB_Rating'] 
    group['normalized'] = (val-minval)/(maxval-minval)
    return group

genres.apply(apply_normal)
```
