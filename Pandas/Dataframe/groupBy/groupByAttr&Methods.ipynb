```python
import pandas as pd
import numpy as np

movies = pd.read_csv("imdb-top-1000.csv")
movies
```

```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
len(movies.groupby('Genre'))
movies['Genre'].nunique()
```

```python
movies.groupby('Genre').size()
```

```python
genres = movies.groupby('Genre')
genres.size()
# genres.first()
# genres.last()
genres.nth(6)#0 based indexing
```

```python
# get grp
genres.get_group('Horror')
```

```python
genres.groups
# har ek group mein jo jo item baitha hai uska index
```

```python
# get a random from each group
genres.sample(2,replace = True)
```
