```python
import pandas as pd
import numpy as np

movies = pd.read_csv('imdb-top-1000.csv')
duo = movies.groupby(['Director','Star1'])# pass a list
duo.size()

duo.get_group(('Aamir Khan','Amole Gupte'))#pass in a tuple
```

```python
# find the most earning actor->director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
actgen = movies.groupby(['Star1','Genre'])
actgen['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
#reset_index() lagane se series gets converted to dataframe
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl
```

```python
# top 10 batsmen in terms off runs scored
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# batsmen with most amount of sixes
six = ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
over = ipl[ipl['over']>15]
group = over[(over['batsman_runs']==4)|(over['batsman_runs']==6)]

group.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batsman'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
def retfunc(batsmen):
    bats = ipl[ipl['batsman']==batsmen]
    return bats.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)

retfunc('V Kohli')
```
