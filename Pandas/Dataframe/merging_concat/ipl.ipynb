```python
import pandas as pd
import numpy as np

delivery =pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')
matches
```

```python
delivery
```

```python
# find top 3 stadiums with highest sixes/match ratio
# match_id ka naam alag hai to 
stad = delivery.merge(matches,how='inner',left_on='match_id',right_on='id')
temp = stad[stad['batsman_runs']==6]
num_sixes= temp.groupby('venue')['venue'].count()
num_matches= matches['venue'].value_counts()
(num_sixes/num_matches).sort_values(ascending=False)
```
