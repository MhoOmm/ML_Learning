```python
import pandas as pd
import numpy as np

# date_time functions

# timestamp objects - timesatmps refer particular instance of time 
pd.Timestamp('2023/12/13')

# variations 
pd.Timestamp('2023-12-13')
# pd.Timestamp('2023, 13, 12') # comma seperated
pd.Timestamp('13th December 2025')

# AM/PM
pd.Timestamp('13th December 2025 5:56AM')
```

```python
# python has a date/time pbject
import datetime as dt
x = pd.Timestamp(dt.datetime(2025,12,13,5,56,00))
# fething data
x.day
x.year
x.month
x.hour
x.minute
```

```python
# # why separate objects to handle data and time when python already has datetime functionality?
# syntax wise datetime is very convenient
# But the performance takes a hit while working with huge data. List vs Numpy Array
# The weaknesses of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.
# The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly.
```

```python
import numpy as np
date = np.array('2015-07-04', dtype=np.datetime64)
date + np.arange(12)
```

```python
# Because of the uniform type in NumPy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large
# Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64
# From a group of these Timestamp objects, Pandas can construct a DatetimeIndex that can be used to index data in a Series or DataFrame
```

```python
# date_range function
pd.date_range(start='2025/12/13',end='2025/12/31',freq ='D')
```

```python
# alternate dates
pd.date_range(start='2025/12/13',end='2025/12/31',freq ='2D')
```

```python
# Business Days - mon - fri
pd.date_range(start='2025/12/13',end='2025/12/31',freq ='B')
```

```python
pd.date_range(start='2025/12/13',end='2025/12/31',freq ='W-FRI')
```

```python
# H - hourly factor
pd.date_range(start='2025/12/13',end='2025/12/14',freq ='7H')
```

```python
pd.date_range(start='2025/12/13',end='2030/12/31',freq ='A')
```

```python
# abhi se leke agle 25 din
pd.date_range(start='2025/12/13',periods=25,freq ='D')
```

```python
# to-datetime function - converts an existing object to datetimeindex object

s = pd.Series(['2025-12-13', '2025-12-14', '2025-12-15', '2025-12-16',
               '2025-12-17', '2025-12-18', '2025-12-19', '2025-12-20',
               '2025-12-21', '2025-12-22'])
# s.str.split('-').str.get(0)
pd.to_datetime(s).dt.year
pd.to_datetime(s).dt.month_name
pd.to_datetime(s).dt.month
pd.to_datetime(s).dt.day

# with errors - ignore this value
pd.to_datetime(s,error='coerce')
```
