```python
import numpy as np 
import pandas as pd
```

```python
data= pd.read_csv('IMDB Dataset.csv')
data.head()
```

Preprocessing Stage

```python
df = data.iloc[1:1500:]
df.shape
```

```python
df.head()
```

Lowercasing

```python
df['review'] = df['review'].str.lower()
df.head()
```

Removing HTML Tags

```python
import re

def remove_html_tagsre(text):
    clean = re.compile(r'<.*?>', re.DOTALL)
    return re.sub(clean, '', text)

df['review'] = df['review'].apply(remove_html_tagsre)
df.head()
```

Removing Punctuation

```python
import string
exclude = string.punctuation
exclude
```

```python
# def remove_punc(text):
#     word = []
#     for i in text.split():
#         if i in exclude:
#             word.append("")
#         else:
#             word.append(i)

# takes a lot of time so we apply this

def removepunc(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
df['review'] = df['review'].apply(removepunc)
df.head()
```

Removing Stopwords

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
print(list(stop_words)) 
```

```python
# def remove_stopwords(text):
#     new_text = []
#     for word in text.split():
#         if word in stopwords.words('english'):
#             new_text.append("")
#         else:
#             new_text.append(word)
    
#     return " ".join(new_text)

# faster  way
from nltk.corpus import stopwords

# Preload stopwords once
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # Split and filter using set membership
    return " ".join(word for word in text.split() if word not in stop_words)
```

```python
df['review'] = df['review'].astype(str).apply(remove_stopwords)
```

```python
df.head()
```

Find out the number of words in the entire corpus

```python
words = []
for review in df['review']:
    for word in review.split():
        words.append(word)
len(words)
```

the total number of unique words(vocabulary) using just python

```python
unique = set(words)
unli = list(unique)
len(unli)
```

Apply bag words and find the vocabulary also find the times each word has occured

```python
from sklearn.feature_extraction.text import CountVectorizer
cnv = CountVectorizer()
bow = cnv.fit_transform(df['review'])
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())

bow.toarray().shape
```

```python
cnv.vocabulary_
```

Apply bag of bi-gram and bag of tri-gram and write down your observation about the dimensionality of the vocabulary

```python
# bi gram
from sklearn.feature_extraction.text import CountVectorizer
cnvbi = CountVectorizer(ngram_range=(2,2)) 
bowbi = cnvbi.fit_transform(df['review'])
bowbi.toarray().shape
```

```python
cnvbi.vocabulary_
```

```python
# tri gram
from sklearn.feature_extraction.text import CountVectorizer
cnvtri = CountVectorizer(ngram_range=(3,3)) 
bowtri = cnvtri.fit_transform(df['review'])
bowtri.toarray().shape
```

```python
cnvtri.vocabulary_
```

Apply tf-idf and find out the idf scores of words, also find out the vocabulary

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfi = TfidfVectorizer(ngram_range=(1,3))
tfibow = tfi.fit_transform(df['review']).toarray()

print(tfibow.shape)
```

```python
tfibow
```

```python
tfi.vocabulary_
```

```python
print(tfi.idf_)
print(tfi.get_feature_names_out())
```
