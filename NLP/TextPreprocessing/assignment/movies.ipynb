```python
import pandas as pd
import numpy as np
data = pd.read_csv("movies.csv")
```

```python
data.head()
```

```python
df = data[['title_x','genres','summary']]
df.head()
```

Preprocessing starts

Lowercasing

```python
df['title_x']=df['title_x'].str.lower()
df['genres']=df['genres'].str.lower()
df['summary']=df['summary'].str.lower()
df.head()
```

Removing Punctuation

```python
import string
punctuations = string.punctuation
print(punctuations)
```

```python
def remove_punc(text):
    newsent = []
    for word in text.split():
        if word in punctuations:
            newsent.append("")
        else :
            newsent.append(word)
    return " ".join(newsent)
```

```python
df['title_x']=df['title_x'].apply(remove_punc)
df['summary']=df['summary'].apply(remove_punc)
```
