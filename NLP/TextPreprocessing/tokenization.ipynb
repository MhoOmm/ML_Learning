Library functions

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize

text = "Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book."

tokens = word_tokenize(text)
print(tokens)
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize

text = "Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book."
tokens2 = sent_tokenize(text)
print(tokens2)
```

```python
from nltk.tokenize import word_tokenize
tx1 = " I have an Ph.d in A.I"
tx2 = " We're here to help! Reach out to us as example@gmail.com"
tx3 = " A 5km ride is going to cost you $10.50"
tx1 = word_tokenize(tx1)
tx2 = word_tokenize(tx2)
tx3 = word_tokenize(tx3) # problem
print(tx1,tx2,tx3)
```

Best Library for tokenization is Spacy

```python
import spacy
nlp = spacy.load('en_core_web_sm')

tx1 = " I have an Ph.d in A.I"
tx2 = " We're here to help! Reach out to us as example@gmail.com"
tx3 = " A 5km ride is going to cost you $10.50"

tx1 = nlp(tx1)
tx2 = nlp(tx2)
tx3 = nlp(tx3)
```

```python
for token in tx1:
    print(token)
```

```python
for token in tx2:
    print(token)
```

```python
for token in tx3:
    print(token)
```
