Removing HTML Tags

```python
corpus = """<!DOCTYPE html>
<html>
<head>
    <title>Sample Document</title>
</head>
<body>

<h1>Welcome to <b>My Blog</b></h1>

<p>Hello there! 👋 This is a <span style="color:red;">simple</span> blog post.</p>

<p>
    I love <i>coding</i>, <strong>music</strong>, and
    <a href="https://example.com">open-source</a> projects.
</p>

<div class="content">
    <h2>Things I Do</h2>
    <ul>
        <li>Write <b>Python</b> code</li>
        <li>Practice <i>Data Structures</i></li>
        <li>Build <span class="highlight">side projects</span></li>
    </ul>
</div>

<div class="ad">
    <p>🔥 Limited Offer! <a href="#">Click here</a> to win prizes!</p>
</div>

<p>
    Thanks for reading!<br>
    <em>Have a great day 😊</em>
</p>

<!-- Footer -->
<footer>
    <p>&copy; 2026 My Blog</p>
</footer>

</body>
</html>"""
```

```python
import re

def remove_html_tagsre(text):
    clean = re.compile(r'<.*?>', re.DOTALL)
    return re.sub(clean, '', text)

corpus = remove_html_tagsre(corpus)
corpus = corpus.lower()
print(corpus)
```

```python
df['Reviews'] = (
    df['Reviews']
    .fillna("")
    .astype(str)
    .str.lower()
    .apply(remove_html_tagsre)
)

df.head()
```

Removing URLS

```python
toy_reviews = [
    "Great product! Check it out at https://example.com/product/123",
    "Worst service ever 😡 visit http://badservice.net for proof",
    "Loved it!!! More details at www.shopnow.in/deals",
    "Support page: https://help.example.org?ref=chat",
    "No links here, just text 😊",
    "Follow us on https://twitter.com/company and http://facebook.com/company",
    "This site (www.fake-site.co.uk) is a scam!!!"
]

import re

def remove_urls(text):
    url_pattern = r'https?://\S+|www\.\S+'
    return re.sub(url_pattern, '', text)

for i in range(len(toy_reviews)):
    toy_reviews[i] = remove_urls(toy_reviews[i])
    
print(toy_reviews)
```

```python
df['Reviews'] = df['Reviews'].apply(remove_urls)
df.head()
```

Removing Stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
print(list(stop_words)) 
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append("")
        else:
            new_text.append(word)
    
    return " ".join(new_text)

remove_stopwords("Hello my name is So  this is OMM TRIPATHI")
```

```python
# df['Reviews'] = df['Reviews'].astype(str).apply(remove_stopwords)
nedf = df.iloc[1:11:]
nedf['Reviews'] = nedf['Reviews'].astype(str).apply(remove_stopwords)
nedf
```

Remove Emojis

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U000024C2-\U0001F251" 
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
text = "Feeling sad now 😢💔"
cleaned = remove_emojis(text)
print(cleaned)
```

```python
nedf['Reviews'] = nedf['Reviews'].astype(str).apply(remove_emojis)
nedf
```
