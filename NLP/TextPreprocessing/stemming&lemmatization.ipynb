```python
from nltk.stem.porter import PorterStemmer
pd = PorterStemmer()
```

```python
def stem_words(text):
    return " ".join([pd.stem(word) for word in text.split()])
```

```python
text = "walks walking walk walked"
stem_words(text)
```

```python
text = """Researchers were analyzing and analyzed large datasets while computers were running faster than ever before. 
The systems had been optimized, optimizing, and are still being optimized by engineers who design and redesign 
intelligent solutions. Users were clicking, clicked, and will click on applications that provide better services, 
serving millions of customers worldwide. Children were playing, played, and are playing happily while their parents 
were watching them closely. Cars are driven, driving, and were driven on roads that had been repaired, repairing, 
and are being repaired by workers. The studies show that better decisions are made when people think carefully, 
thought deeply, and are thinking logically. Businesses are growing, grew, and have grown rapidly in markets where 
competition increases, increased, and is increasing continuously.
"""

stem_words(text)
```

```python
import nltk 
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
wdlm = WordNetLemmatizer()

sentence = """Researchers were analyzing and analyzed large datasets while computers were running faster than ever before. 
The systems had been optimized, optimizing, and are still being optimized by engineers who design and redesign 
intelligent solutions. Users were clicking, clicked, and will click on applications that provide better services, 
serving millions of customers worldwide"""

punctuations = string.punctuation
word_box = word_tokenize(sentence)

for word  in word_box:
    if word in punctuations:
        word_box.remove(word)
        
for word in word_box:
    print(word,"|||||",wdlm.lemmatize(word,pos='v'))
```
