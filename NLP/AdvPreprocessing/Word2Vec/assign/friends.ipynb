```python
import numpy as np 
import pandas as pd
```

```python
import gensim 
import os
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

with open("Friends_Transcript.txt",encoding="utf-8", errors="ignore") as f:
    data = f.read()
    
tokens = sent_tokenize(data)
    
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    return " ".join(word for word in text.split() if word not in stop)

newtokens = []

for sentence in tokens:
    newtokens.append(remove_stopwords(sentence))
    
```

```python
corpus = []

for sentence in newtokens: 
    corpus.append(simple_preprocess(sentence))
    
# corpus
```

```python
model = gensim.models.Word2Vec(
    window = 50,
    min_count = 2,
    vector_size= 200
)
```

```python
model.build_vocab(corpus)
```

```python
model.train(corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar("chandler")
```
