```python
import numpy as np
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
a2
```

```python
# scalar operations
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
a1 = a1**2 # scalar ops
a1 = a1 > 8 # relational ops
print(a1)
```

```python
# vector OPs (two arrays of same shape can be operated together)
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
a1+a2
```

```python
import numpy as np
a2 = np.arange(12,24).reshape(3,4)
np.transpose(a2)
a2.T
```

```python
a3 = np.arange(12,24).reshape(2,2,3)
print(a3)
np.ravel(a3)
```

```python
# stacking & splitting

# Horizontal stacking
a1 = np.arange(12).reshape(3,4)
a2  = np.arange(12,24).reshape(3,4)
np.hstack((a1,a2))
np.vstack((a1,a2))
```

```python
# Splitting
a1 = np.arange(16).reshape(4,4)
a2  = np.arange(12,28).reshape(4,4)
print(np.hsplit(a1,2))
print(np.vsplit(a2,2))
```
