```python
# har row ka ops
import numpy as np
a1 = np.random.random((3,4))
a1  = np.round(a1*100)
print(a1)
# 0-> col ans 1->row
np.max(a1 , axis = 0)
np.min(a1,axis = 1)

# har col ya row ka prod
np.sum(a1,axis=0)
np.prod(a1,axis=1)
```

```python
# statistical ops
import numpy as np
a1 = np.random.random((3,4))
a1  = np.round(a1*100)
print(a1)
np.mean(a1)
np.mean(a1,axis=0)

np.median(a1)
np.var(a1) #variance
np.std(a1) #standard deviation
```

```python
import numpy as np
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12).reshape(4,3)
print(a1)
print(a2)
np.dot(a1,a2)
```

```python
# log and exponents
np.log(a1)
np.exp(a1)
```
