```python
import numpy as np
a1 = np.random.randint(1,100,15)
a1 = np.sort(a1)
print(a1)

```

```python
# 2-d array

a2 = np.random.randint(1,100,15).reshape(3,5)
print(a2)
print("\n")
b = np.sort(a2,axis=1)
print(b)
```

```python
# np.append

# np.concatenate
c = np.random.randint(1,100,15).reshape(3,5)
d = np.random.randint(1,100,15).reshape(3,5)
print(np.concatenate((c,d),axis=0))
```

```python
a = np.array([1,1,1,2,3,4,5,3,4,5,5])
print(a)

print(np.unique(a))
```

```python
a1 = np.random.randint(1,50,10)
print(a1)
print(a1.shape)
print(np.expand_dims(a1,axis=0).shape)
```

```python
a1 = np.random.randint(1,100,10)
print(a1)
print(a1[a1>50])
print(np.where(a1>50))
```

```python
# can change array with condition

# np.where(condn,true,false)
print(np.where(a1>50,0,a1))
print(np.where(a1>50,a1,0))
```

```python
# argmax :- greatest element at which position
# argmin :- smallest element at which position
np.argmax(a1)

d = np.random.randint(1,100,15).reshape(3,5)
print(d)
np.argmax(d)
```

```python
# cumsum 
# cumprod
print(a1)
# print(np.cumsum(a1))
# print(np.cumprod(a1))

d = np.random.randint(1,100,15).reshape(3,5)
print(d)
print(np.cumsum(d,axis=1))
```

```python
# for histrogram
print(a1)
np.histogram(a1,(10,20,30,40,50,60,70,80,90))
```

```python
# for pearson - corelation
print(a1)
salary = np.array([20000,30000,40000,50000,60000])
# exp = np.array([1,2,3,4,5]) +ve corelation
# exp = np.array([5,4,3,2,1]) -ve corelation
exp = np.array([8,3,5,1,5])

print(np.corrcoef(salary,exp))
```

```python
# isin - tells us that multiple elements are present in  an array of not
b = np.array([10,2,23,86,37])
print(np.isin(b,a1))
```

```python
# flip- mirror image of the object keeping the shape intact
c = np.arange(1,10)
print(c)
print(np.flip(c))
```

```python
c = np.arange(1,11).reshape(2,5)
print(c)
print(np.flip(c,axis=0)) #row-wise
print(np.flip(c)) # both columnwise and rowwise
```
