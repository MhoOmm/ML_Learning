```python
import numpy as np
a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
print(a1)
print(a2)
print(a3)
```

```python
# indexing
print(a1[-1])
print(a2[0][2])
print(a3[1][1][1])
```

```python
a1
print(a1[2:6:1])
```

```python
print(a2)
print(a2[0,:])
print(a2[:,2])
```

```python
print(a2)
# print(a2[0:2,0:])
print(a2[::2,::3 ])
# print(a2[::2,1::2])
# print(a2[1:2:,::3])
print(a2[0:2:1,1::1])
```

```python
a3 = np.arange(27).reshape(3,3,3)
# print(a3)
# print(a3[2,1::,1::])
print(a3[0::2,0:1:2,::2])
```
