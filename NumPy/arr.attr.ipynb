```python
import numpy as np
a1 = np.arange(10)
a2 = np.arange(16,dtype = int).reshape((4,4))
a3 = np.arange(18).reshape((2,3,3))
# print(a1)
# print(a2)
# print(a3)
```

```python
# ndim = gives us the dimension of the numpy array
a1.ndim
a2.ndim
a3.ndim
```

```python
# gives us the shape array
a1.shape
a2.shape
a3.shape
```

```python
# size : gives us the total number of elements
print(a3.size)
```

```python
# itemsize: gives us the space taken by a datatype

# dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```
