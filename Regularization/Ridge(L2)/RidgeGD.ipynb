```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
X,y = load_diabetes(return_X_y=True)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
class meraRdg :
    def __init__(self,learning_rate,alpha,epochs):
        self.epochs = epochs
        self.lr = learning_rate
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
    
    def fit(self,X_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        w = np.insert(self.coef_,0,self.intercept_)
        X_train = np.insert(X_train,0,1,axis=1)
        
        for i in range(self.epochs):
            derivative  = np.dot(X_train.T,X_train).dot(w) - np.dot(X_train.T,y_train) + w*self.alpha
            w = w - self.lr * derivative
            
        self.intercept_ = w[0]
        self.coef_ = w[1:]
     
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
reg = meraRdg(epochs=500,alpha=0.001,learning_rate=0.005)
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```
