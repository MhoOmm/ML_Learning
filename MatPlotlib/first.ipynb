```python
# adding labels
batsman = pd.read_csv('sharma-kohli.csv')
print(batsman)

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma Vs Virat Kohli Career comp')
plt.xlabel('Year')
plt.ylabel('Runs in Hundreds')
```

```python
# changing colour and other changes
batsman = pd.read_csv('sharma-kohli.csv')
print(batsman)

plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dashed',marker='P')
plt.plot(batsman['index'],batsman['RG Sharma'],color='gold',marker='D')
plt.title('Rohit Sharma Vs Virat Kohli Career comp')
plt.xlabel('Year')
plt.ylabel('Runs in Hundreds')

# linestyle - dashed,dotted,dashdot,solid
# linewidth = 2,3
# to see the points in a 2d plot 
# marker size = 5,10
```

```python
# adding label
batsman = pd.read_csv('sharma-kohli.csv')
print(batsman)

plt.plot(batsman['index'],batsman['V Kohli'],color='purple',linestyle='dashed',marker='P',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='gold',marker='D',label='Rohit')
plt.title('Rohit Sharma Vs Virat Kohli Career comp')
plt.xlabel('Year')
plt.ylabel('Runs in Hundreds')
plt.legend()
```

```python
df = pd.read_csv('batsman_season_record.csv')
print(df)
# plt.bar(df['batsman'],df['2015'])
# plt.bar(df['batsman'],df['2016'])
# plt.bar(df['batsman'],df['2016'])

# multiple barplots

plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2016'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```
