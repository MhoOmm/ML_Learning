```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

plt.style.use('dark_background')
```

```python
iris = pd.read_csv('iris.csv')

iris['Species'] = iris['Species'].replace({'Iris-virginica':0,'Iris-setosa':1,"Iris-versicolor":2})
iris.sample(10)
```

```python
# to colour map the scatter plots
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
# to change colour add a cmap attribute {winter,viridis,summer etc}
```

```python
# increasing the plot size
plt.figure(figsize=(7,4))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
# (width,length)
```

```python
# sample annotation example
r1 = [1,2,4,5,6]
r2 = [3,5,6,2,6]

plt.scatter(r1,r2)
plt.text(1,3,'point1')
plt.text(2,5,'point2')
```

```python
df = pd.read_csv('batter.csv')
sample = df.head(100).sample(25,random_state=5)

plt.figure(figsize=(15,7))

plt.scatter(sample['avg'],sample['strike_rate'])
# plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])

# annotations - kaise pata chalega konsa dot konda batter hai
for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
    
# how to plot horizontal and vertical lines on a graph
plt.axhline(130,color='yellow')
plt.axvline(30,color='red')
```
