```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

batter = pd.read_csv('batter.csv')

fig,ax = plt.subplots()
ax.scatter(batter['avg'],batter['strike_rate'],color='red',marker='+')
ax.set_xlabel('Batsman Average')
ax.set_ylabel('Strike Rate')
ax.set_title('Batsman avg vs Strike Rate')
```

```python
#  to plot multiple plots 
# nrows * ncols amount of graphs can be done
fig,ax = plt.subplots(nrows=2,ncols=1)
ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red',marker='+')
ax[0,1].scatter(batter['avg'],batter['runs'],color='red',marker='+')
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1) # 2 rows 2 columns 1st graph
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'])
```
